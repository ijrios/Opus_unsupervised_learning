```python
import scipy.io as sio
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```


# Primer base de datos

```python
data=sio.loadmat('/content/databases_opus_quattour/Aggregation.mat')
dato = data['Aggregation']
```

### Implementaci√≥n del Kmeans para la primera base de datos



```python
# Por defecto usa la distancia euclidiana
km = KMeans(
    n_clusters=3, init='random',
    n_init=10, random_state=0
)

y_km = km.fit_predict(dato)
# plot the clusters
plt.scatter(dato[:, 0], dato[:, 1], cmap='viridis',c=y_km)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

list_k = list(range(2, 10))

for n_clusters in list_k:
    clusterer = KMeans(n_clusters=n_clusters)
    preds = clusterer.fit_predict(dato)
    centers = clusterer.cluster_centers_

    score = silhouette_score (dato, preds)
    print ("For n_clusters =" + str(n_clusters) + " silhouette score is " + str(score))
```

```python
# 2 clusters
clusterer_2 = KMeans(n_clusters=2,n_init=10)
preds_kmeans_2 = clusterer_2.fit_predict(dato)
# 3 clusters
clusterer_3 = KMeans(n_clusters=3,n_init=10)
preds_kmeans_3 = clusterer_3.fit_predict(dato)
# 4 clusters
clusterer_4 = KMeans(n_clusters=4,n_init=10)
preds_kmeans_4 = clusterer_4.fit_predict(dato)
# 5 clusters
clusterer_5 = KMeans(n_clusters=5,n_init=10)
preds_kmeans_5 = clusterer_5.fit_predict(dato)
# 6 clusters
clusterer_6 = KMeans(n_clusters=6,n_init=10)
preds_kmeans_6 = clusterer_6.fit_predict(dato)
# 7 clusters
clusterer_7 = KMeans(n_clusters=7,n_init=10)
preds_kmeans_7 = clusterer_7.fit_predict(dato)
# 8 clusters
clusterer_8 = KMeans(n_clusters=8,n_init=10)
preds_kmeans_8 = clusterer_8.fit_predict(dato)
# 9 clusters
clusterer_9 = KMeans(n_clusters=9,n_init=10)
preds_kmeans_9 = clusterer_9.fit_predict(dato)
```

### Implementaci√≥n del Spectral Clustering para la primera base de datos

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# create spectral clustering object
spectral = SpectralClustering(n_clusters=5, affinity='rbf', gamma=1)

# fit and predict clusters
labels = spectral.fit_predict(dato)

# plot the clusters
plt.scatter(dato[:, 0], dato[:, 1], c=labels, cmap='viridis')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

list_k = list(range(2, 10))

for n_clusters in list_k:
    clusterer = SpectralClustering(n_clusters=n_clusters,assign_labels='discretize')
    preds = clusterer.fit_predict(dato)
    centers = clusterer.labels_

    score = silhouette_score (dato, preds)
    print ("For n_clusters =" + str(n_clusters) + " silhouette score is " + str(score))
```

```python
# 2 clusters
SC_clusterer_2 = SpectralClustering(n_clusters=2,assign_labels='discretize')
preds_SpectralClustering_2 = SC_clusterer_2.fit_predict(dato)
# 3 clusters
SC_clusterer_3 = SpectralClustering(n_clusters=3,assign_labels='discretize')
preds_SpectralClustering_3 = SC_clusterer_3.fit_predict(dato)
# 4 clusters
SC_clusterer_4 = SpectralClustering(n_clusters=4,assign_labels='discretize')
preds_SpectralClustering_4 = SC_clusterer_4.fit_predict(dato)
# 5 clusters
SC_clusterer_5 = SpectralClustering(n_clusters=5,assign_labels='discretize')
preds_SpectralClustering_5 = SC_clusterer_5.fit_predict(dato)
# 6 clusters
SC_clusterer_6 = SpectralClustering(n_clusters=6,assign_labels='discretize')
preds_SpectralClustering_6 = SC_clusterer_6.fit_predict(dato)
# 7 clusters
SC_clusterer_7 = SpectralClustering(n_clusters=7,assign_labels='discretize')
preds_SpectralClustering_7 = SC_clusterer_7.fit_predict(dato)
# 8 clusters
SC_clusterer_8 = SpectralClustering(n_clusters=8,assign_labels='discretize')
preds_SpectralClustering_8 = SC_clusterer_8.fit_predict(dato)
# 9 clusters
SC_clusterer_9 = SpectralClustering(n_clusters=9,assign_labels='discretize')
preds_SpectralClustering_9 = SC_clusterer_9.fit_predict(dato)
```
